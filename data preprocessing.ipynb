```python
import pandas as pd

train  = pd.read_csv("raw_train_data.csv")
test = pd.read_csv("raw_test_data.csv")
```

```python
train
```

```python
print train.columns
print(len(train.columns))
```

#### Data needs to be scaled to a small range like 0 to 1

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
```

```python
scale_train = scaler.fit_transform(train)
scale_test = scaler.transform(test)
```

```python
# Print out the adjustment that the scaler applied to the total_earnings column of data
print("total_earnings values were scaled by multiplying by {:.10f} and adding {:.6f}".format(scaler.scale_[8], scaler.min_[8]))
```

<strong> Creating DataFrame from scale Data </strong>

```python
scale_train_df = pd.DataFrame(scale_train, columns = train.columns.values)
scale_test_df = pd.DataFrame(scale_test, columns = test.columns.values)
```

### SAVING DATAFRAME AS CSV

```python
scale_train_df.to_csv('sales_data_training_scaled.csv' , index = False)
scale_test_df.to_csv('sales_data_test_scaled.csv' , index = False)
```
